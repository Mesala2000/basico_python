```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv("959_elecciones_2023_filtrado.csv")

# Seleccionar columnas
columnas_seleccionadas = ["PARTIDO","NUM_VOTOS","PORCENTAJE"]
df_agrupado = df.groupby(['PARTIDO']).sum()

# Tratar de dibujar las columnas seleccionadas
fig, ax = plt.subplots(figsize=(8,8))

# Extraer datos para el grfico de pastel
partidos = df_agrupado.index
num_votos = df_agrupado['NUM_VOTOS']

# Dibujar el grfico de pastel
ax.pie(num_votos, labels=partidos, autopct='%1.1f%%',
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, startangle=140)

# Configuraciones adicionales del grfico
ax.set_title('Porcentaje de votos por partido')
ax.axis('equal')  # Hacer que el grfico de pastel sea circular

plt.show()

```
