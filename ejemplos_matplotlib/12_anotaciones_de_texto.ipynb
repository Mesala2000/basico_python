```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.subplot()

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             ) #situación, color, tamaño, etc. de la flecha y el texto 

plt.ylim(-2, 2)#límites del eje y de la grafica, de -2 hasta 2
plt.show()
```
