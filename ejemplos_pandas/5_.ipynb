```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
from scipy.stats import norm, binom

"""
https://en.wikipedia.org/wiki/Quantile

"""

# make data
str_data = r"""date,x,y
2016-12-31,0,2
2017-01-01,1,2
2017-01-02,2,2 
2017-01-03,3,2
2017-01-04,4,2
2017-01-05,5,2
2017-01-06,6,3
2017-01-07,7,3
2017-01-08,8,4
2017-01-09,9,4
"""
#instanciamos un df de tipo pandas y leemos el archivo str_data línea por línea
df = pd.read_csv(io.StringIO(str_data))
df.set_index(['date'], inplace=True) #le añadimos un index, que va a ser la columna date

df['x'] = df['x'].astype(float) #a la columna x determinamos que los valores que va a contener sean de tipo float.
df['y'] = df['y'].astype(float)
#crea una copia exaxta de nuestro df. si la manipulamos la df original no se modifica.
df_copy = df.copy()

# replace - 1
#imprime el df original
print(df)
mask = (df['x'] > 0.0) & (df['x'] < 3.0) #crea un booleano con condiciones y devuelve true si se cumple, y false si no se cumplen
df.loc[mask, 'x'] = -1 #accede a los datos de df que coinciden con mask y sustituye los datos de x por -1
print(df)

# replace - 2
#imprime el df_copy
print(df_copy)
df_copy['x'] = np.where((df_copy['y'] == 3.0) | (df_copy['y'] == 4.0), np.nan, df_copy['x']) # a los valores de la columna x de df_copy les pasa una condicion. Si se cumple 
                                                                                             # al registro le asigna el valor nan, sino nada.
print(df_copy)
```
