```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#creamos n√∫meros aleatorios con numpy a tres tipos de valores (sunny, rain cloudy). la muestra es 1000. Tb se especifica la probabilidad de cada valor.
data = np.random.choice(['sunny', 'rain', 'cloudy'], 1000, p=[1 / 4.0, 1 / 4.0, 1 / 2.0])
#creamos un data frame con pandas y le rellenamos con los datos de data
df = pd.DataFrame(data=data, columns=['weather'])

df['weather'].value_counts().plot(kind='bar')#En la columna weather del df se van acumulando los vaqlores. Se plotean en un grafico de barras.
#imprimimos por consola
plt.show()
```
